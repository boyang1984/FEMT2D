```python
import pygimli as pg
import matplotlib.pyplot as plt
import pygimli.meshtools as mt
import os
from os import system
import numpy as np
```

```python
def callTriangle(filename,
                 quality=33, 
                 triangle='triangle',
                 verbose=True):
    
    filebody = filename.replace('.poly', '')

    syscal = triangle + ' -pq' + str(quality)
    syscal += 'Aa ' + filebody + '.poly'
 
    if verbose:
        print(syscal)

    system(syscal)
```

```python
world = mt.createWorld(start=[-4e5, -4e5], 
                       end=[4e5, 4e5], 
                       layers=[0, 18e3, 26e3], 
                       area=[0, 1e7, 1e6, 0],
                       marker=[1, 2, 3, 4],
                       worldMarker=False)
```

```python
geom = world
```

```python
ax, _ = pg.show(geom, 
                showNodes=False, 
                boundaryMarker=False)
ax.set_ylim(ax.get_ylim()[::-1]);
```

```python
mt.exportPLC(geom,
             '../meshes/layered.poly', 
             float_format='.8e')
```

```python
callTriangle('../meshes/layered.poly', 
             quality=34.2,
             verbose=False)
```

```python
mesh = mt.createMesh(geom, quality=34.2)
ax, _ = pg.show(mesh)
ax.set_ylim(ax.get_ylim()[::-1]);
```
