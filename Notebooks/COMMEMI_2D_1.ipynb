```python
import pygimli as pg
import matplotlib.pyplot as plt
import pygimli.meshtools as mt
import os
from os import system
import numpy as np
```

```python
def callTriangle(filename,
                 quality=33, 
                 triangle='triangle',
                 verbose=True):
    
    filebody = filename.replace('.poly', '')

    syscal = triangle + ' -pq' + str(quality)
    syscal += 'Aa ' + filebody + '.poly'
 
    if verbose:
        print(syscal)

    system(syscal)
```

```python
world = mt.createWorld(start=[-1e5, -1e5], 
                       end=[1e5, 1e5], 
                       layers=[0], 
                       area=[1e7, 0],
                       marker=[1, 2],
                       worldMarker=False)
```

```python
block = mt.createRectangle(start=[-0.5e3, 0.25e3],
                           end=[0.5e3, 3.25e3],
                           marker=3,  
                           markerPosition=[0, 1e3], 
                           area=1e3, boundaryMarker=10)
```

```python
geom = world + block
```

```python
ax, _ = pg.show(geom, 
                showNodes=False, 
                boundaryMarker=False)
ax.set_ylim(ax.get_ylim()[::-1]);
```

```python
mt.exportPLC(geom,
             'meshes/commemi2d1.poly', 
             float_format='.8e')
```

```python
callTriangle('meshes/commemi2d1.poly', 
             quality=34.2,
             verbose=False)
```

```python
mesh = mt.createMesh(geom, quality=34.2)
ax, _ = pg.show(mesh)
ax.set_ylim(ax.get_ylim()[::-1]);
```
